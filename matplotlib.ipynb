```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
y
```

```python
x
```

```python
plt.plot(x,y,'r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph')
```

```python
plt.subplot(2,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(2,2,4)
plt.plot(y, x, 'g*-');
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.5,0.7])
axes.plot(x,y,'b')

axes.set_xlabel('--x--')
axes.set_ylabel('--y--')
axes.set_title('--title--')
```

```python
fig2=plt.figure()
axes1=fig2.add_axes([0.1,0.1,0.9,0.9])
axes2=fig2.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('big X')
axes1.set_ylabel('big Y')
axes1.set_title('Big')

axes2.plot(y,x,'r')
axes2.set_xlabel('small x')
axes2.set_ylabel('small y')
axes2.set_title('small')
```

SUBPLOTS

```python
f,a=plt.subplots()
a.plot(x,y,'*')
a.set_xlabel('x')
a.set_ylabel('y')
a.set_title('Sub Graph')
```

```python
f1,a1=plt.subplots(nrows=1,ncols=2)
```

```python
for i in a1:
  i.plot(x,y,'green')

  
f1
```

2D

```python
f2,a2=plt.subplots(nrows=2,ncols=2)
k=1
for i in a2:
  for j in i:
    if k%2==0:
      j.plot(x,y,'p')
      k+=1
    else:
      j.plot(y,x,'r')
      k+=1
```

```python
fig1=plt.figure(figsize=(8,2),dpi=80)
a1=fig1.add_axes([0.1,0.1,1,1])
a1.plot(x,y,'r', label="x vs y")
a1.plot(y,x,'b', label="y vs x")
a1.legend()
```

```python
fq,aq=plt.subplots(figsize=(12,3))
aq.plot(x,y,'r', label="x vs y")
aq.plot(y,x,'b', label="y vs x")
aq.legend(x,loc=2)
```
