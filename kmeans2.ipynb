```python
import pandas as pd
```

```python
ds=pd.read_csv('ds/College_Data')
```

```python
ds
```

```python
ds=ds.drop('Unnamed: 0',axis=1)
```

```python
ds
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=2)
```

```python
km.fit(ds.drop('Private',axis=1))
```

```python
km.labels_
```

```python
km.cluster_centers_
```

```python
def conv(k):
    if k=='Yes':
        return 1
    else:
        return 0
```

```python
ds['cluster']=ds['Private'].apply(conv)
```

```python
ds
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ds['cluster'],km.labels_)
```
